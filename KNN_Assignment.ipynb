Implement K-NN Classifier for classification of any dataset of your choice.

```python
# a. Load the dataset
iris_data = load_iris()
features = iris_data['data']
labels = iris_data['target']
```

```python
# b. Split dataset into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.2, random_state=42
)
```

```python
# c. Initialize variables for storing accuracy scores
k_values = list(range(1, 21))
accuracy_scores = {}

# Iterate through various values of k and store accuracy
for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(train_features, train_labels)
    predictions = model.predict(test_features)

    accuracy = accuracy_score(test_labels, predictions)
    accuracy_scores[k] = accuracy

    # Display accuracy and confusion matrix for k=3
    if k == 3:
        print(f"Accuracy for k=3: {accuracy:.2f}")
        print(f"Confusion Matrix for k=3:\n{confusion_matrix(test_labels, predictions)}")
```

```python
# d. Plotting the accuracy scores for different k-values
plt.figure(figsize=(8, 6))
plt.plot(k_values, list(accuracy_scores.values()), marker='o', color='b', linestyle='--')
plt.title('K-NN: Accuracy vs Value of K', fontsize=14)
plt.xlabel('Value of K', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.grid(True)
plt.show()

# Find the optimal k with maximum accuracy
optimal_k = max(accuracy_scores, key=accuracy_scores.get)
max_accuracy = accuracy_scores[optimal_k]

print(f"Optimal value of k: {optimal_k}, Maximum Accuracy: {max_accuracy:.2f}")
```
